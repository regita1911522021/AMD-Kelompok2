```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('swalayan.csv', header = None)
ds
```

```python
transactions = []
for i in range(0, 7501):
  transactions.append([str(ds.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidence  = [result[2][0][2] for result in results]
    lift        = [result[2][0][3] for result in results]
    
    return list(zip(lhs, rhs, supports, confidence, lift))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Produk 1', 'Produk 2', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame.drop(['Confidence','Lift'],axis=1).sort_values(by='Support',ascending=True)
```
