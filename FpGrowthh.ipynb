```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Grape', 'Orange', 'Banana', 'Kiwi', 'Pear', 'Mango'],
           ['Durian', 'Orange', 'Banana', 'Kiwi', 'Pear', 'Mango'],
           ['Grape', 'Apple', 'Kiwi', 'Pear'],
           ['Grape', 'Strowberry', 'Corn', 'Kiwi', 'Mango'],
           ['Corn', 'Orange', 'Orange', 'Kiwi', 'Coconut', 'Pear']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```
